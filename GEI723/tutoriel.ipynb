```python
from brian2 import *
# Paramètres simples pour un modèle de neurone
tau = 10*ms
eqs = '''
dv/dt = (1-v)/tau : 1
'''
# Configuration du neurone
G = NeuronGroup(1, eqs, method='exact')
M = StateMonitor(G, 'v', record=0)
G.v = 0
# Exécuter la simulation
run(100*ms)
# Tracer le résultat
plot(M.t/ms, M.v[0])
xlabel('Temps (ms)')
ylabel('Potentiel de membrane (v)')
show()
```
